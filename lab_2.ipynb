```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k = 0.0001
  return k*x
t = [2003, 2005, 2009, 2010, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
solve = odeint(dydt, 1063800, t=t)
solve = solve.flatten()
%matplotlib inline 

v = [1098100, 1093000, 1093699, 1095909, 1143458, 1156201, 1170000, 1182221, 1192036, 1198858, 1202371, 1200719, 1196680, 1187960, 1179288]

plt.plot(t, solve, label = r"x′=kx")
plt.plot(t, v)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
a=-100000
b=0.5
def dydt(x,t):
  return k(x)*x*0.00355
def k(x):
  return a/x+b
t = [2003, 2005, 2009, 2010, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
solve = odeint(dydt, 1063800, t=t)
solve = solve.flatten()
%matplotlib inline 

v = [1098100, 1093000, 1093699, 1095909, 1143458, 1156201, 1170000, 1182221, 1192036, 1198858, 1202371, 1200719, 1196680, 1187960, 1179288]

plt.plot(t, solve, label = r"x'=k(x)*x")
plt.plot(t, v)
plt.grid(True)
plt.legend()
plt.show()
```
